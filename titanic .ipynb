```python
female=data[data.Sex=='female']
```

```python
plt.bar(female['Name'], female['Survived'])
plt.title('titanic')
plt.ylabel('PassengerId')
plt.gcf().set_size_inches(20, 10)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.barh(female['Name'], female['Survived'],color='GREEN')
plt.title('titanic')
plt.xlabel('PassengerId')
plt.gcf().set_size_inches(20, 10)
#plt.xticks(rotation=240)
plt.show()
```

```python
Name=data['Age'].value_counts().head(10).values
name=data['Age'].value_counts().head(10).index
plt.gcf().set_size_inches(10, 5)
plt.pie(name, labels=name, autopct="%1.1f%%", wedgeprops={'edgecolor':'RED'}, explode=[0.5,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05])
plt.title('titanic')
plt.show()
```

```python
plt.bar(data['Pclass'], data['Age'],color=(data['Age']>40).map({True:'#DBB40C', False:'RED'}))
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.gcf().set_size_inches(8,2)
plt.xticks(rotation=90)
plt.show()
```
